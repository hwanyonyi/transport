```python
import apache_beam as beam
import datetime

def process_data(element):
    # Convert the dictionary to a CSV string
    csv_string = ','.join(str(v) for v in element.values())
    return csv_string

# Create a pipeline
with beam.Pipeline() as p:
    # Read data from the "to_hive" file
    data = (
        p
        | beam.io.ReadFromText("data/to_hive-00000-of-00001")
        | beam.Map(eval)  # Convert each line (dictionary) to a Python dictionary
    )

    # Convert dictionary data to CSV strings
    csv_data = data | beam.Map(process_data)

    # Write CSV data to a text file
    csv_data | beam.io.WriteToText("data.csv")

# Now you have the processed data in the "data.csv" file

# Load data into Hive
# Use Hive's LOAD DATA command to load data from the "data.csv" file into your Hive table
# Replace "your_table" with the actual table name
# Execute this command using the Hive command-line interface or a client library

# Example Hive command:
# LOAD DATA LOCAL INPATH '/path/to/data.csv' OVERWRITE INTO TABLE your_table;
```
